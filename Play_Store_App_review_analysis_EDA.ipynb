### 3.3 Changing Datatypes into required formats

As price column contain $ symbol and datatype is object so it's mandatory to remove that symbol and convert the values into float datatype

````python
# Price column contain $ symbol which is object datatype
df1['Price'].unique()
````

````python
# replace $ symbol from price column with empty string
df1['Price'] = df1['Price'].astype(str).str.replace('$', '')
````

````python
# Converting Price column datatype from string to float
df1['Price'] = df1['Price'].astype(float)
````

Installs column values carry '+' symbol and 'comma' as the datatype is object
in order to remove those symbols and convert object datatype into interger by using following codes:

````python
# As Installs column having '+' and ',' symbol 
# Replace it with empty string
df1['Installs']=df1['Installs'].astype(str).str.replace('+','')
df1['Installs']=df1['Installs'].astype(str).str.replace(',','')
````

````python
# Converting Installs column values from string to int datatype
df1['Installs']=df1['Installs'].astype(int)
````

````python
#  to convert object datatype into int datatype
df1['Reviews'] = df1['Reviews'].astype(int)
````

As Size column having 'M' or 'k' at the end by removing and converting size column values in MB

````python
# defining function to convert column data values into MB.
def convert_size(size):
    if 'M' in size:
        # If size is already in megabytes, remove the 'M' and return the size
        return float(size[:-1])
    elif 'k' in size:
        # If size is in kilobytes, divide by 1024 and return the size in megabytes
        return float(size[:-1])/1024
    elif 'Varies with device' in size:
        # If size varies with device, return NaN
        return float('NaN')
    else:
        # If size is in bytes, divide by 1024^2 and return the size in megabytes
        return float(size[:-1])/1024**2

# Apply the convert_size function to the size column
df1['Size'] = df1['Size'].apply(convert_size)
````

As Last updated column carries values in datetime format

````python
#  We have to convert 'last updated' column datatype to datetime format

df1["Last Updated"] = [datetime.strptime(i, '%B %d, %Y') for i in df1["Last Updated"]]
````

````python
# Chart - 6.1 visualization code bar chart

# plt.figure(figsize=(10, 8))
colors = plt.cm.viridis(np.linspace(0, 1, len(category_count['Category'])))
plt.bar(category_count['Category'],category_count['App'], color=colors)
plt.xticks(rotation=90)
plt.xlabel('Category')
plt.ylabel('Number of Apps')
plt.title('Most Popular Category on the Play Store(in terms of count)')
plt.show()
````

````python
# Chart - 6.2 visualization code donut chart


labels = category_count['Category']
values = category_count['App']

fig = go.Figure(data=[go.Pie(labels=labels, values=values,
                             textinfo='percent', insidetextorientation='radial')])

fig.update_traces(hole=0.4, hoverinfo='label+percent+name', 
                  marker=dict(line=dict(color='#000000', width=1)))

fig.update_layout(title='Percentage distribution of apps in each Category', 
                  showlegend=True, 
                  width=1200, 
                  height=800,
                  margin=dict(l=50, r=50, b=100, t=100, pad=4),
                  paper_bgcolor='rgb(243, 243, 243)',
                  plot_bgcolor='rgb(243, 243, 243)',
                  font=dict(family='Arial, sans-serif', size=15, color='#000000'))

fig.show()
````

````python
# Chart - 6.3 visualization code bar chart
# Data for the bar chart
categories = category_count['Category']
num_apps = category_count['App']

# Calculate percentages
total_apps = sum(num_apps)
percentages = [(num_app/total_apps)*100 for num_app in num_apps]

# Create bar chart
fig = plt.figure(figsize=(10, 8))
colors = plt.get_cmap('tab20').colors
plt.barh(categories, num_apps, color=colors)

# Add percentages as text on each bar
for i in range(len(categories)):
    plt.text(num_apps[i]+2, i, f'{percentages[i]:.2f}%', va='center')

# Add axis labels and title
plt.xlabel('Number of Apps')
plt.ylabel('Categories')
plt.title('Percentage of Apps in Each Category')

# Show the chart
plt.show()
````
